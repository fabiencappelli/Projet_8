```python
import os
import random
import shutil

splits = ['train', 'val']

masksSourceFolder = '/content/drive/MyDrive/Colab Notebooks/Projet 8/P8_Cityscapes_gtFine_trainvaltest/gtFine8'
imagesSourceFolder = '/content/drive/MyDrive/Colab Notebooks/Projet 8/P8_Cityscapes_leftImg8bit_trainvaltest/leftImg8bit'

for split in splits:
  # Dossiers d'origine
  images_dir=os.path.join(imagesSourceFolder, split)
  masks_dir=os.path.join(masksSourceFolder, split)

  # Dossiers de destination
  sample_images_dir = os.path.join('/content/drive/MyDrive/Colab Notebooks/Projet 8/dataset_sample/images', split)
  sample_masks_dir = os.path.join('/content/drive/MyDrive/Colab Notebooks/Projet 8/dataset_sample/masks', split)

  # Crée les dossiers si besoin
  os.makedirs(sample_images_dir, exist_ok=True)
  os.makedirs(sample_masks_dir, exist_ok=True)

  # Liste des fichiers
  image_files = sorted([f for f in os.listdir(images_dir) if f.endswith('leftImg8bit.png')])
  prefixes = [f.replace('leftImg8bit.png', '') for f in image_files]

  # On prend 15 % d'images au hasard
  random.seed(34)
  sample_size = int(len(prefixes) * 0.15)
  sampled_prefixes = random.sample(prefixes, sample_size)

  print(f"Nombre d'images sélectionnées : {sample_size}")

  # On copie les fichiers images et masques associés
  for prefix in sampled_prefixes:
      img_file = prefix + 'leftImg8bit.png'
      mask_file = prefix + 'gtFine_labelIds.png'

      # Vérifie l’existence
      img_path = os.path.join(images_dir, img_file)
      mask_path = os.path.join(masks_dir, mask_file)
      if os.path.exists(img_path) and os.path.exists(mask_path):
          shutil.copy2(img_path, os.path.join(sample_images_dir, img_file))
          shutil.copy2(mask_path, os.path.join(sample_masks_dir, mask_file))
      else:
          print(f"Manquant : {img_path} ou {mask_path}")
```
